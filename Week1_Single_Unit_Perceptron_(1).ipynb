a. Design a single unit perceptron for classification of a linearly separable binary dataset (placement.csv)
without using pre-defined models. Use the Perceptron() from sklearn.

```python

# Single unit perceptron
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
df=pd.read_csv('Placement.csv')
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
p = Perceptron()
p.fit(X,y)
print(p.coef_)
print(p.intercept_)
z=p.score(X,y)
print("accuracy score is",z)
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from mlxtend.plotting import plot_decision_regions

# XOR Data
xor_data = pd.DataFrame()
xor_data['input1'] = [1, 1, 0, 0]
xor_data['input2'] = [1, 0, 1, 0]
xor_data['output'] = [0, 1, 1, 0]

# Input and Output
X = xor_data.iloc[:, 0:2].values
y = xor_data.iloc[:, -1].values

# Define and train a multilayer perceptron
mlp = MLPClassifier(hidden_layer_sizes=(4,), activation='relu', max_iter=1000, random_state=1)
mlp.fit(X, y)

# Print weights and bias
print("Weights:", mlp.coefs_)
print("Biases:", mlp.intercepts_)
print("Accuracy:", mlp.score(X, y))

# Plot decision boundary
plot_decision_regions(X, y, clf=mlp, legend=2)
plt.title("XOR Gate using Multilayer Perceptron")
plt.xlabel("input1")
plt.ylabel("input2")
plt.show()
```
